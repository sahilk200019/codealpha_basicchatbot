```python
import re
import random

class RuleBot:
    negative_responses = (
        "I am  sorry, I can not  help with that. ",
        "sorry to say , I don't have the information you're looking for.",
        "I am sorry and apolise , but I cannot assist with this  request.",
        "i am unable to make request this time.",
        "Sorry, I don't have the answer to that question.",
        "I'm afraid I can't provide the help you're seeking.",
        "That request is outside of my capabilities.",
        "I'm sorry, but I'm unable to fulfill that request.",
        "Unfortunately, I can't assist you with that issue.",
        "I'm unable to handle that query right now."
    )

    exit_commands = (
        "exit",
        "quit",
        "bye",
        "goodbye",
        "see you",
        "later",
        "end",
        "stop",
        "close",
        "terminate",
        "done"
    )
    def __init__(self):
        self.responses = {
            "greeting": r".*\b(Assalmalikum e|hello|Namste|good\s*morning|good\s*evening)\b.*",
            "help": r".*\b(help|assist|support)\b.*",
            "information_request": r'.*\b(data\s*science)\b.*',
            "farewell": r".*\b(goodbye|bye|see\s*you|talk\s*to\s*you\s*soon|catch\s*you\s*later)\b.*",
            "thank_you": r'.*\b(thank\s*you|thanks|great|good)\b.*',
            "joke_request": r'.*\b(joke)\b.*',
            "weather_inquiry": r'.*\b(weather)\b.*',
            "intro_to_bot": r'.*\b(about\s*yourself|who\s*are\s*you)\b.*',
            "no_match_found": r'.*'
        }
    random_questions = (
       "How can I assist you today?",
    "Is there something specific you need help with?",
    "What would you like to know more about?",
    "Do you need assistance with anything?",
    "How can I help make things easier for you?",
    "Is there a particular issue you're facing that I can help with?",
    "What information are you looking for?",
    "Can I help you find what you're looking for?",
    "What can I do to assist you?",
    "Do you need help navigating something?"
    )

   

    def greeting(self):
        self.name = input("What is your name?\n")
        will_help = input(f"Hi {self.name}, I am Support chat-bot. How may i help you?\n")
        if will_help.lower() in self.negative_responses:
            print("Ok, have a nice day!")
            return
        self.chat()

    def make_exit(self, reply):
        for command in self.exit_commands:
            if command in reply:
                print("Okay, have a nice day!")
                return True
        return False

    def chat(self):
        reply = input(random.choice(self.random_questions) + "\n").lower()
        while not self.make_exit(reply):
            reply = input(self.match_reply(reply) + "\n").lower()

    def match_reply(self, reply):
        for key, value in self.responses.items():
            found_match = re.match(value, reply)
            if found_match:
                if key == 'information_request':
                    return self.information_request()
                elif key == 'weather_inquiry':
                    return self.weather_inquiry()
                elif key == 'joke_request':
                    return self.joke_request()
                elif key == 'intro_to_bot':
                    return self.intro_to_bot()
                elif key == 'thank_you':
                    return self.thank_you()
                elif key == 'farewell':
                    return self.farewell()
        return self.no_match_found()

    

    def weather_inquiry(self):
        responses = (
            "I'm sorry, I can't provide real-time weather updates right now. Please check your local weather service for the latest information.",
            "I don't have access to current weather data. Please use a weather app or website for up-to-date information.",
            "Unfortunately, I can't check the weather for you. Try asking your smartphone or a weather service instead."
        )
        return random.choice(responses)
    
    def information_request(self):
        responses = (
            "Data Science is a field that combines domain expertise, programming skills, and knowledge of mathematics and statistics to extract meaningful insights from data.",
            "Data Science involves using algorithms, data analysis, and machine learning to understand and analyze actual phenomena with data.",
            "Data Science is an interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from noisy, structured, and unstructured data."
        )
        return random.choice(responses)
    def joke_request(self):
        responses = (
             "टीचर: बताओ अगर धरती से सूरज गायब हो जाए तो क्या होगा?\nस्टूडेंट: फिर तो तामसिक हो जाएगा!",
    "मोहन: तुमसे शादी करने के लिए मैं बहुत प्यार करता हूँ।\nसपना: कितनी बार कहा है, लव का जवाब लव से देते हैं, स्टेपलर से नहीं!",
    "बॉस: तुमने आज ऑफिस क्यों नहीं आए?\nकर्मचारी: सर, मुझे तो पता ही नहीं था कि आज ऑफिस है!",
    "मास्टर: अगर तुम्हारे पास 10 सेब हैं और तुम्हारे दोस्त ने 5 ले लिए, तो तुम्हारे पास कितने सेब बचेंगे?\nछात्र: मेरे पास 10 सेब ही रहेंगे, क्योंकि मेरे दोस्त ने बिना पूछे ही ले लिए!",
    "बच्चा: मम्मी, मैं बड़ा होकर क्या बनूँगा?\nमम्मी: बेटा, तुम बड़ा होकर खुदा बनोगे, तुम्हारे सामने कोई नहीं टिकेगा!"
        )
        return random.choice(responses)

   

    def intro_to_bot(self):
        responses = (
            "Hi, I'm your friendly chatbot here to assist you with any questions you have!",
            "Hello! I'm a chatbot designed to help you with various inquiries. How can I assist you today?",
            "Greetings! I'm an AI chatbot, here to provide you with information and support. What can I do for you today?"
        )
        return random.choice(responses)
    def thank_you(self):
        responses = (
            "You're welcome! If you have any other questions, feel free to ask.",
            "No problem at all! I'm here to help.",
            "Anytime! Let me know if there's anything else you need."
        )
        return random.choice(responses)
    def no_match_found(self):
        responses = (
            "I'm not sure how to respond to that.",
            "Can you please rephrase your question?",
            "I didn't quite understand that. Could you ask in a different way?"
        )
        return random.choice(responses)

Bot = RuleBot()
Bot.greeting()
```
